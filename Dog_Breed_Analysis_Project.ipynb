```python
def height_min(x):
  print (len(x))
  if len(x) == 5:
   return int(x[:2])
  elif len(x)== 4:
    print(x)
    return int(x[:1])
  elif len(x) == 3:
    return int(x[:1])


def height_max(x):
  if len(x) == 5:
    return int (x[-2:])
  elif len(x) == 4:
    print(x)
    return int (x[-2:])
  elif len(x) == 3:
    return int (x[-1])
```

```python
df['min_height'] = df['height (in)'].apply (height_min)
df['max_height'] = df['height (in)'].apply (height_max)
df_a['min_height'] = df_a['height (in)'].apply (height_min)
df_a['max_height'] = df_a['height (in)'].apply (height_max)
```

```python
df['min_height'].plot(kind='hist', title = 'Distribution of Minimum Dog Height', figsize=(8,4))
```

```python
df['max_height'].plot(kind='hist', title='Distribution of Maximum Dog Height',
                      figsize=(8,4),)
```

```python
df['longevity (yrs)']
```

```python
def longevity_min(x):
  print(len(x))
  if len(x) == 5:
    return int(x[:2])
  elif len(x) == 4:
    print(x)
    return int(x[:1])
  elif len(x) == 3:
    return int(x[:1])


def longevity_max(x):
  if len(x) == 5:
    return int(x[-2:])
  elif len(x) == 4:
    print(x)
    return int(x[-2:])
  elif len(x) == 3:
    return int(x[-1])
```

```python
df['min_longevity'] = df['longevity (yrs)'].apply(longevity_min)
df['max_longevity'] = df['longevity (yrs)'].apply(longevity_max)
df_a['min_longevity'] = df_a['longevity (yrs)'].apply(longevity_min)
df_a['max_longevity'] = df_a['longevity (yrs)'].apply(longevity_max)
```

```python
df['min_longevity'].plot(kind='hist', title='Distribution of Minimum Dog Longevity',
                      figsize=(8,4),)
plt.show()
df['min_longevity'].value_counts (normalize=True)
```

```python
df['max_longevity'].plot(kind='hist', title='Distribution of Maximum Dog Longevity',
                      figsize=(8,4))
plt.show()
df['max_longevity'].value_counts (normalize=True)
```
